```python
#  3. Divide List Elements: Define a function divide_elements(L, n) to divide each element of list L by
#  number n. Handle ZeroDivisionError and TypeError

def divide_elements(L, n):
     for i in range(len(L)):
        try:
          
            L[i]=L[i]//n
        except (ZeroDivisionError, TypeError):
            print("Check for 0 division or Type.....")
            return
     return L


L=[1,2,0,1]
print(divide_elements(L,0))
```

```python
#  8. Dictionary Update: Write a function that updates a dictionary only if the key already exists,
#  otherwise raises a custom exception KeyNotFoundError.

def update(D,key):
    try:
        if key in D:
            value = input("Enter the new value: ")
            D[key] = value
            print("Dictionary updated:", D)
        else:
            raise KeyError("KeyNotFoundError: The key does not exist in the dictionary.")
    except KeyError as e:
        print(e)


D = {"name":"Abhay", "Branch":"Ec", "ROll":2411085}
keyy = input("Enter the key to update: ")
update(D, keyy)
```

```python
# 13. Nested Try Example: Write code where you take a list and divide each element by 2 inside a try,
#  and handle outer exception for TypeError if the input is not a list

try:
    mylist = [1, 2, 3, 4, 5]
    for i in range(len(mylist)):
        try:
            mylist[i] = mylist[i] // 2
        except ZeroDivisionError as e:
            print("Error: ", e)
    print("Updated List:", mylist)
except TypeError:
    print("Input must be a list.")
    
```

```python
# 14. Using Else in Exception: Write code to open a file, print its contents, and display 'File read
#  successfully' in the else block. Handle FileNotFoundError.

try:
    with open("sample.txt", "r") as file:
        content = file.read()
        print(content)
except FileNotFoundError:
    print("File not found.")
else:
    print("File read successfully.")
    
```
