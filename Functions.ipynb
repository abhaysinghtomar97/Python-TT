```python
#WAP to take your Full name as input from the user and only print the initials of the name.


def initials(name):
    sp = name.split()
    initials = ""
    for part in sp:
        initials += part[0]
    return initials

name = input("Enter name: ")
print(initials(name))

```

```python
def printName(first,last, initials=False):
    if initials:
        print(f"{first[0]}.{last[0]}")
    else:
        print(f"{first} {last}")

printName("Abhay", "Singh")
```

```python
#WAP To findout area of circile take r as input from the user and set pi as default value.

def area(r, pi=3.14):
    return pi * r * r

radius = float(input("Enter radius : "))
print(area(radius))
```
